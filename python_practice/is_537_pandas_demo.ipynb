```python
df_passenger = df['passenger_info']
df_names = df_passenger['Name'].str.split(' ', expand=True)
full_df = pd.merge(df_passenger, df_names, how='outer', left_index=True, right_index=True)
```

```python
full_df
```

```python
df_survival = df['survival_info'].rename(columns={'PID':'PassengerId'})
```

```python
all_df = pd.merge(full_df, df_survival, how='left', on='PassengerId')
```

```python
all_df
```

```python
all_df['Entrance'] = np.where(all_df['Main Deck Entrance'] == True, 'this is a true', 'this is a false') 
all_df['Entrance']
```

```python
all_df['Entrance'] = np.where(all_df['Main Deck Entrance'] == True, 'Main Deck Entrance',
                              (np.where(all_df['South Entrance'] == True, 'South Entrance',
                                        (np.where(all_df['North Entrance'] == True, 'North Entrance', np.nan)))))
```
